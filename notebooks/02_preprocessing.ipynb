# Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../dataset/heart_disease.csv')
df.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder

numeric_features = df.select_dtypes(include='float64').columns
categorical_features = df.select_dtypes(include='str').columns[:-1]
```

```python
X = df.drop('Heart Disease Status', axis=1)
y = df['Heart Disease Status']
```

```python
# Pipeline for saving imputed data into a new csv file

num_clean_pipeline = Pipeline(steps=[
    ("imputer", KNNImputer(n_neighbors=5))
])

cat_clean_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy='most_frequent'))
])

clean_preprocessor = ColumnTransformer(transformers=[
    ("num", num_clean_pipeline, numeric_features),
    ("cat", cat_clean_pipeline, categorical_features)
])

combined_features = list(numeric_features) + list(categorical_features)

X_clean = clean_preprocessor.fit_transform(X)
y_clean = LabelEncoder().fit_transform(y)

clean_df = pd.DataFrame(X_clean, columns=combined_features)
clean_df['Heart Disease Status'] = y_clean

float_cols = clean_df.select_dtypes(include='object').columns[:9]

for col in float_cols:
    clean_df[col] = pd.to_numeric(clean_df[col], errors="coerce")

clean_df.to_csv('../dataset/cleaned_dataset.csv', index=False)
print("Cleaned Dataset Saved")
```

```python
# Pipeline to impute and encode the data

num_pipeline = Pipeline(steps=[
    ("imputer", KNNImputer(n_neighbors=5)),
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy='most_frequent')),
    ("encoder", OneHotEncoder())
])

preprocessor = ColumnTransformer(transformers=[
    ("num", num_pipeline, numeric_features),
    ("cat", cat_pipeline, categorical_features)
])
```

```python
X_transformed = preprocessor.fit_transform(X)
y_transformed = LabelEncoder().fit_transform(y)
```
